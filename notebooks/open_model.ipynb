```python
import tensorflow.keras as keras
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
lstm_model = keras.models.load_model("../models/lstm_ckpt/lstm64_lstm64_batch256.h5")
```

```python
log_data = pd.read_csv('../models/lstm_log/lstm64_lstm64_b256.log', sep=',', engine='python')
```

```python
def visualize_loss_MAE(history):
    val_loss = history.history["val_loss"]
    val_mae = history.history["val_mean_absolute_error"]
    epochs = range(len(val_loss))
    fig, ax = plt.subplots(1,2, figsize=(10,4))
    ax[0].plot(epochs, val_loss, label="Validation loss")
    ax[1].plot(epochs, val_mae, label="Validation MAE")
    ax[0].set_title('Validation loss')
    ax[1].set_title('Validation Mean Absolute Error')
    ax[0].set_xlabel('epochs')
    ax[1].set_xlabel('epochs')
    ax[0].set_ylabel('loss')
    ax[1].set_ylabel('MAE')
```
